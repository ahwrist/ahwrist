```python
#import pandas and datetime
import pandas as pd
import datetime
```

```python
#read in data set from Kaggle People HR Analytics Repository: https://www.kaggle.com/HRAnalyticRepository/employee-attrition-data/version/2
df = pd.read_csv(r"/Users/andrewwrist/Documents/Projects/MFG10YearTerminationData_raw.csv")
df.head()
```

```python
#rename columns to more friendly names
df.columns = df.columns.str.replace('_key','') #remove _key from end of column names
df = df.rename(columns = {'EmployeeID': 'emplid', 'recorddate':'status_date', 'birthdate':'birth_date','orighiredate':'hire_date', 'terminationdate':'term_date', 'department_name':'dept_descr', 'gender_short':'gender', 'gender_full':'gender_descr','termreason_desc':'term_reason','termtype_desc':'term_type','STATUS_YEAR':'status_year','STATUS':'status','BUSINESS_UNIT':'business_unit'}) #rename column names to more friendly/consistent names
df
```

```python
#take out 1/1/00 dates
df = df.replace('1/1/00','')
```

```python
#Expand yearly status dates to monthly
df2 = df #create copy data frame
df2['status_date'] = pd.to_datetime(df2['status_date']) #convert status_date to datetime for indexing
df2 = df2.set_index('status_date').groupby('emplid',group_keys=False).resample('M').bfill().reset_index() #set index by status date and group by employee id
df2

```

```python
#export to CSV
df2.to_csv('HR_Data_by_Month_End_Date2.csv', index = False)
```
